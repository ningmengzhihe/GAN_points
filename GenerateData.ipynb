```python
import numpy as np
import pandas as pd
import os
```

```python
n_sample = 1001
n_highdim = 992
np.random.seed(42)
```

```python
if not os.path.exists('./data'):
    os.makedirs('./data')
```

```python
ma = np.random.randint(1, 150, size=(1,n_sample))[0]
a = np.random.randint(-20, 120, size=(1,n_sample))[0]
b = np.random.randint(0, 160, size=(1,n_sample))[0]
```

```python
df_name = pd.DataFrame(data=np.vstack([ma, a, b]).T, columns=['ma', 'a', 'b'])
df_name['name'] = [f'ma{ma[i]}a{a[i]}b{b[i]}' for i in range(0, n_sample)]
```

```python
df_name
```

```python
df_name.to_csv('./data/df_name_points.csv', index=False)
```

```python
data_high = np.random.randn(n_sample, n_highdim)
df_highdim = pd.DataFrame(data=data_high)
df_highdim['name'] = df_name['name']
df_highdim
```

```python
df_highdim.to_csv('./data/df_highdim_points.csv', index=False)
```
